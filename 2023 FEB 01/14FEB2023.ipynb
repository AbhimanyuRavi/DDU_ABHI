```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Car_sales.csv')
df.sample(30)
```

### DROPING NA VALUES

```python
df.dropna(inplace=True)
```

```python
df.info()
```

### PLOTS ON NUMERICAL VALUE COLUMNS

#### HISTOGRAM

```python
df.Horsepower.plot.hist(figsize=(10,5), cmap='binary_r');
```

### BOX PLOT

```python
df.Sales_in_thousands.plot.box(figsize=(4,3), cmap = 'viridis');
```

```python
df.Engine_size.plot.box(figsize=(10,9),cmap='gnuplot');
```

### SUBPLOTS

```python
df.plot.box(figsize=(28,8), subplots=True);
```

```python
df.plot.hist(figsize=(8,24),subplots = True);
```

### LINE CHARTS

```python
df.sort_values(by='__year_resale_value').plot.line('__year_resale_value','Price_in_thousands', figsize=(6,3),cmap='inferno');
```

### BAR PLOT

#### To create the below shown bar chart we have to create the frequency table,frequency table can be made using the value_counts(),method

```python
df.Manufacturer.value_counts().plot.bar(figsize=(6,2), cmap= 'nipy_spectral');
```

```python
df.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3));
```

### AREA CHART

```python
df.plot.area('Manufacturer','Price_in_thousands',figsize=(6,3), cmap='seismic');
```

### HORIZONTAL BAR CHART

```python
df.sort_values(by='Price_in_thousands')[:8].plot.barh('Model','Power_perf_factor',figsize=(6,4));
```

### SCATTER PLOT

```python
df.sort_values(by='Price_in_thousands')[:5].plot.scatter('Latest_Launch','Price_in_thousands',figsize=(10,3));
```

```python
df.plot.box(figsize=(22,8));
```

```python
df.sort_values(by='Engine_size',ascending=False)[:10].Engine_size.plot.pie();
```

### KDE PLOT

```python
df.plot.kde(figsize=(10,4));
```

#### subplots

```python
df.plot.kde(figsize=(12,10),subplots=True);
```

### Hexbin plot

```python
df.plot.hexbin('Curb_weight','Fuel_capacity', gridsize=15, cmap='nipy_spectral')
```
